Importing dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
#sets a style for plotting
```

plotting a vector

```python
plt.quiver(0,0,1,1, scale_units= 'xy', angles= 'xy', scale=1)
plt.xlim(0,3)
plt.ylim(0,3)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles= 'xy', scale =1, color= 'b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,-3,-6, scale_units='xy', angles= 'xy', scale =1, color= 'g')
plt.quiver(0,0,4,5, scale_units='xy', angles= 'xy', scale =1, color= 'b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Vector addition

```python
vector1= np.asarray([0,0,2,-3])

vector2= np.asarray([0,0,3,3])

vector3= vector1+vector2

print(vector3)
```

```python
plt.quiver(0,0,2,-3, color= 'r', scale_units= 'xy', angles= 'xy', scale=1)
plt.quiver(0,0,3,3, color= 'y', scale_units= 'xy',angles= 'xy', scale=1)
plt.quiver(0,0,5,0, color= 'b', scale_units= 'xy', angles= 'xy',scale=1)
plt.xlim(-7,7)
plt.ylim(-7,7)
plt.show()
```

Vector Substraction

```python
vector1= np.asarray([0,0,2,-3])

vector2= np.asarray([0,0,3,3])

vector3= vector1-vector2

print(vector3)
```

```python
plt.quiver(0,0,2,-3, color= 'r', scale_units= 'xy', angles= 'xy', scale=1)
plt.quiver(0,0,3,3, color= 'y', scale_units= 'xy',angles= 'xy', scale=1)
plt.quiver(0,0,-1,-6, color= 'b', scale_units= 'xy', angles= 'xy',scale=1)
plt.xlim(-7,7)
plt.ylim(-7,7)
plt.show()
```

Vector Multiplication

```python
vector1= np.asarray([0,0,2,-3])

vector2= np.asarray([0,0,3,3])

vector3= vector1*2
vector4= vector2*-0.5

print(vector3, vector4)
```

```python
plt.quiver(0,0,2,3, color= 'r', scale_units= 'xy', angles= 'xy', scale=1)
plt.quiver(0,0,3,3, color= 'y', scale_units= 'xy',angles= 'xy', scale=1)
plt.quiver(0,0,4,6, color= 'b', scale_units= 'xy', angles= 'xy',scale=1)
plt.quiver(0,0,-1.5,-1.5, color= 'g', scale_units= 'xy', angles= 'xy',scale=1)
plt.xlim(-7,7)
plt.ylim(-7,7)
plt.show()
```
