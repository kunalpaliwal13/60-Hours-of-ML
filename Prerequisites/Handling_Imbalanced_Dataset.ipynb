```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
credit_data= pd.read_csv('/content/creditcard.csv')
```

```python
credit_data.dropna(axis=0, inplace = True)
```

```python
credit_data.isnull().any()
```

```python
credit_data['Class'].value_counts()
#highly imbalanced
```

```python
legit = credit_data[credit_data.Class==0]
fraud = credit_data[credit_data.Class==1]
```

```python
print(legit.shape, fraud.shape)
```

**UnderSampling**

```python
legit_sample = legit.sample(n=142)
#we take 142 data sample from legit , this is called undersampling
```

```python
legit_sample.shape
```

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
```
